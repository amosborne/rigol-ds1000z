```python
from rigol_ds1000z import Rigol_DS1000Z, process_display
```

```python
with Rigol_DS1000Z() as oscope:
    oscope.ieee(rst=True)
    oscope.autoscale()
    oscope.timebase(main_scale=200e-6)
    oscope.single()

    display = oscope.display()
    process_display(display, show=True, filename="display_capture.png")
```
