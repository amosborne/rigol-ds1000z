```python
with Rigol_DS1000Z() as oscope:
    oscope.autoscale()
    oscope.timebase(main_scale=200e-6)
    oscope.single()

    waveform = oscope.waveform(source=2)
    xdata, ydata = process_waveform(
        waveform, show=True, filename="waveform_capture.csv"
    )
```
